```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
fid_report_path = "../reports/validation_fid.csv"

df_fid = pd.read_csv(fid_report_path)
df_fid.drop(columns=["Unnamed: 0"], inplace=True)
df_fid["epoch"] = df_fid["epoch"].replace({"final":199})

df_fid["epoch"] = df_fid["epoch"].apply(lambda x: int(float(x)))

df_fid
```

```python
plt.figure(figsize=(16, 10))

plt.plot(df_fid["epoch"].tolist(), df_fid["fid_x"].tolist(), label="Horse", zorder=5, color="#B20600")
plt.plot(df_fid["epoch"].tolist(), df_fid["fid_y"].tolist(), label="Zebra", zorder=5, color="#00092C")

ax = plt.gca()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.legend(fontsize=20)
plt.xlabel("Epoch", fontsize=30)
plt.ylabel("FID", fontsize=30)

plt.xlim([0,200])
plt.ylim([0,400])

plt.xticks(size=20)
plt.yticks(size=20)

plt.grid(zorder=0, alpha=0.3)

plt.savefig("../plots/training/FID.png")
```
